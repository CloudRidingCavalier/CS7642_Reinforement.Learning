```python
for episode in range(atEstablishment.shape[0]):
    print "Now is episode: ", episode
    #print 'Now H is:', total_H
    
    prediction_pool = []
    for policy_index in total_H:
        
        policy = policy_Array[policy_index]
        isFighting = 0
        
        if len(atEstablishment[episode]) != len(policy):
            print("Something wrong!")
        
        y_hat = np.dot(atEstablishment[episode], policy)
        
        if y_hat == 1:
            isFighting = 1
        
        prediction_pool.append(isFighting)
            
    if len(set(prediction_pool)) == 1:
        final_output.append(prediction_pool[0])
    else:
        final_output.append(-1)
        
        removed_indexes = []
        for policy_check_index in total_H:
            policy_check = policy_Array[policy_check_index]
            predict_label = 0
            if np.dot(atEstablishment[episode], policy_check) == 1:
                predict_label = 1
            if int(predict_label) != int(fightOccurred[episode]):
                removed_indexes.append(policy_check_index)
        for removed_index in removed_indexes:
            total_H.remove(removed_index)
        
```
